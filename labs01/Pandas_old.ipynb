### Indeksowanie

```python
import string
litery = list(string.ascii_uppercase)
dane3 = pd.Series(losowe, index=litery)
print(dane3.head())
```

```python
print(dane3['E'])
print(dane3[['P', 'Y', 'T']])
print(dane3[0:7])
```

### Mapowanie

```python
def cube(x):
    return str(x)
print(dane3.map(cube).head())
print(dane3.head())
```

### DataFrame (ramka danych)

```python
wielkie = list(string.ascii_uppercase)
male = list(string.ascii_lowercase)
surowe = list(zip(male, wielkie))

dane = pd.DataFrame(surowe)
print(dane.head())
```

```python
dane.columns = ["małe", "wielkie"]
print(dane.head())
```

```python
for i in dane.iterrows():
    print(i[1]["małe"])
```
